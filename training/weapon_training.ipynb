```python
from ultralytics import YOLO
import torch

def main():
    # -------------------------------
    # Configuration
    # -------------------------------
    MODEL_NAME = "yolov8s.pt"   # best for small datasets
    DATA_YAML = r"/content/drive/MyDrive/Weapon_Model/dataset/data.yaml"
    EPOCHS = 300
    IMG_SIZE = 416
    BATCH_SIZE = 10
    DEVICE = "cuda" if torch.cuda.is_available() else "cpu"

    # -------------------------------
    # Load pretrained YOLOv8 model
    # -------------------------------
    model = YOLO(MODEL_NAME)

    # -------------------------------
    # Train the model
    # -------------------------------
    model.train(
        data=DATA_YAML,
        epochs=EPOCHS,
        imgsz=IMG_SIZE,
        batch=BATCH_SIZE,
        device=DEVICE,
        optimizer="AdamW",     # stable for small datasets
        lr0=2e-4,
        pretrained=True,
        augment=True,
        project= "/content/drive/MyDrive/Weapon_Model/runs",
        name="yolov8n_weapon_detection",
        exist_ok=True
    )

    print("✅ Training completed successfully.")

if __name__ == "__main__":
    main()
```

```python
from ultralytics import YOLO
import os

def main():
    # -------------------------------
    # Paths
    # -------------------------------
    MODEL_PATH = "/content/drive/MyDrive/Weapon_Model/runs/yolov8s_weapon_detection/weights/best.pt"
    VIDEOS_FOLDER = "/content/drive/MyDrive/Weapon_Model/test_data/test_video.mp4"   # folder jisme images hain
    CONF_THRESHOLD = 0.25

    # -------------------------------
    # Load trained model
    # -------------------------------
    model = YOLO(MODEL_PATH)

    # -------------------------------
    # Run inference on image folder
    # -------------------------------
    results = model(
        source=VIDEOS_FOLDER,
        conf=CONF_THRESHOLD,
        save=True,
        project="/content/drive/MyDrive/Weapon_Model",  # parent folder
        name="predicted_video",  # folder jisme results directly save hon
        exist_ok=True
    )

    print("✅ Inference completed successfully.")
    print("📂 Results saved in: /content/drive/MyDrive/Fire_Smoke_model/Test data/predicted_images")

if __name__ == "__main__":
    main()
```
